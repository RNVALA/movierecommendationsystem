```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
movies.dropna(inplace=True)
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
#stamming
#word like love loving loves so after applying stamming we willget ans as love lov love
#!pip install nltk
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
        
```

```python
new_df['tags'].apply(stem)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
new_df['tags'][1]
```

```python
#so we will do text vectorization we will use bag of word technique there are so many technique that we are usingexample are wordtovac exc
#ba gofword
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
cv.fit_transform(new_df['tags']).toarray().shape
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
len(cv.get_feature_names())
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors).shape
```

```python
similarity=cosine_similarity(vectors)
```

```python
sorted(similarity[0],reverse=True)#similarity of 0 index 
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
#here enumerate say the index of movie
#herelamda x:x[1]will sort on the basis  of second column
#1:6 will print onlly first five sort no
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]#it will find index of a particular movie
    distances = similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
